```python
## In Matplotlib, we can draw multiple graphs in a single plot in two ways. One is by using subplot() function ##
## here we are using the subplot this parameters allow us to add keywords to each subplot and its default value is None ##

axs = dataset.plot.area(figsize=(20,15), subplots=True)
```

```python
 sns.displot(dataset)
```

```python
### BAR GRAPH ###
#y = ['FACEBOOK,INSTAGRAM,TWITTER,QUORA,LINKEDIN,REDDIT,SNAPCHAT,YOUTUBE,TIKTOK,EMAIL,OTHER']
#y = dataset.FACEBOOK
plt.figure(figsize=(30,20))
#print(y)
plt.bar(dataset.FACEBOOK, dataset.DATE, label="FACEBOOK",color="c",width=.5)
plt.bar(dataset.REDDIT, dataset.DATE, label="REDDIT",color="g",width=.5)
plt.legend()
plt.show()
```

```python
### SCATTER GRAPH ###

plt.figure(figsize=(50,30))
plt.scatter(dataset.FACEBOOK, dataset.DATE, label="FACEBOOK",color="c")
plt.scatter(dataset.REDDIT, dataset.DATE, label="REDDIT",color="g")
plt.legend()
plt.show()
```

```python
## STACK/AREA PLOT ##

plt.figure(figsize=(50,50))
plt.plot([],[],color='c', label='APP', linewidth=5)
plt.plot([],[],color='g', label='FACEBOOK', linewidth=5)
plt.stackplot(dataset.DATE,dataset.FACEBOOK,dataset.REDDIT,colors=['c','g'])
plt.legend()
plt.show()
```
